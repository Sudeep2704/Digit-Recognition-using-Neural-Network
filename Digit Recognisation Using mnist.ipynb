```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("mnist_train.csv")


print("Dataset shape:", df.shape)  


labels = df.iloc[:, 0].values  
pixel_values = df.iloc[:, 1:].values  


print("Features shape:", pixel_values.shape)  
print("Labels shape:", labels.shape)  


normalized_pixels = pixel_values / 255.0


lowest_value = pixel_values.min()
highest_value = pixel_values.max()
print("Lowest pixel value before normalization:", lowest_value)
print("Highest pixel value before normalization:", highest_value)


manual_normalized_pixels = (pixel_values - 0) / (255 - 0)


print("Are both normalizations equal?", np.allclose(normalized_pixels, manual_normalized_pixels))


images = normalized_pixels.reshape(-1, 28, 28)


fig, axes = plt.subplots(2, 5, figsize=(10, 5))
axes = axes.flatten()

for i in range(10):
    axes[i].imshow(images[i], cmap='gray')
    axes[i].set_title(f"Label: {labels[i]}")
    axes[i].axis("off")
    
    
    print(f"Matrix for Image {i+1} (Label: {labels[i]}):")
    print(images[i])

plt.tight_layout()
plt.show()
```
